```python
import pandas as pd
```

```python
df = pd.read_csv('hour.csv')
df
```

```python
df['day_night'] = df['hr'].apply(lambda x: 'day' if 6 <= x <= 18 else 'night')
```

```python
df.drop(['instant', 'casual', 'registered'], axis=1, inplace=True)
```

```python
df['dteday'] = pd.to_datetime(df.dteday)
df
```

```python
df['season'] = df.season.astype('category')
```

```python
df['holiday'] = df.holiday.astype('category')
df['weekday'] = df.weekday.astype('category')
df['weathersit'] = df.weathersit.astype('category')
df['workingday'] = df.workingday.astype('category')
df['mnth'] = df.mnth.astype('category')
df['yr'] = df.yr.astype('category')
df['hr'] = df.hr.astype('category')
```

```python
df.drop(columns=['dteday'], inplace=True)
df
```

```python
# Create interaction features
df['temp_hum_interaction'] = df['temp'] * df['hum']
df['temp_windspeed_interaction'] = df['temp'] * df['windspeed']
df['hum_windspeed_interaction'] = df['hum'] * df['windspeed']
df
```

```python
# Separating features and target variable
X = df.drop(columns=['cnt']) # Features
y = df['cnt'] # Target
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import TargetEncoder, MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
# Numerical features
numerical_features = ['temp', 'hum', 'windspeed', 'temp_hum_interaction', 'temp_windspeed_interaction', 'hum_windspeed_interaction']
numerical_pipeline = Pipeline([
('imputer', SimpleImputer(strategy='mean')), # Impute missing values with mean
('scaler', MinMaxScaler()) # Normalize using MinMaxScaler
])
```

```python
# Transforming above
X[numerical_features] = numerical_pipeline.fit_transform(X[numerical_features])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Ensure all column names in X_train are strings
X_train.columns = X_train.columns.astype(str)

# Instantiate the Linear Regression model
model = LinearRegression()

# Fit the model to the training data
model.fit(X_train, y_train)

# Make predictions on the training set (you can also use X_test for testing)
y_pred = model.predict(X_train)

# Evaluate the model
mse = mean_squared_error(y_train, y_pred)
r2 = r2_score(y_train, y_pred)

print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared: {r2:.4f}")
```
