```python
import numpy as np
import pandas as pd
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from category_encoders import TargetEncoder

# Custom Linear Regression Model
class SimpleLinearRegression(BaseEstimator, RegressorMixin):
    def __init__(self):
        self.intercept_ = None
        self.coef_ = None

    def fit(self, X, y):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept term
        theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
        self.intercept_ = theta_best[0]
        self.coef_ = theta_best[1:]
        return self

    def predict(self, X):
        X_b = np.c_[np.ones((X.shape[0], 1)), X]  # Add intercept term
        return X_b.dot(np.r_[self.intercept_, self.coef_])

# Load your dataset into a DataFrame named 'data'
data = pd.read_csv('hour.csv')

# Create additional interaction columns
data['day_night_flag'] = np.where(data['hr'] < 6, 0, 1)  # Day/Night flag
data['temp_hum_interaction'] = data['temp'] * data['hum']  # Interaction between temp and humidity
data['temp_windspeed_interaction'] = data['temp'] * data['windspeed']  # Interaction between temp and windspeed
data['hum_windspeed_interaction'] = data['hum'] * data['windspeed']  # Interaction between humidity and windspeed

# Define your features and target variable
features = ['season', 'yr', 'mnth', 'hr', 'holiday', 'weekday', 'workingday',
            'weathersit', 'temp', 'atemp', 'hum', 'windspeed', 'day_night_flag',
            'temp_hum_interaction', 'temp_windspeed_interaction',
            'hum_windspeed_interaction']
target = 'cnt'

# Check if all features are present in the DataFrame
missing_features = [f for f in features if f not in data.columns]
if missing_features:
    raise ValueError(f"Missing columns in the DataFrame: {missing_features}")

# Select features (X) and target variable (y)
X = data[features].values
y = data[target].values

# Define numerical and categorical pipelines
numerical_features = ['temp', 'atemp', 'hum', 'windspeed']  # Example numerical features
categorical_features = ['season', 'weathersit', 'day_night_flag']  # Example categorical features

numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('target_encoder', TargetEncoder())  # Using TargetEncoder
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_pipeline, numerical_features),
        ('cat', categorical_pipeline, categorical_features)
    ]
)

# Calculate the Mean Squared Error (MSE)
mse = np.mean((y - y_pred) ** 2)

# Calculate R-squared (R²)
total_variance = np.var(y) * len(y)
r2 = 1 - (np.sum((y - y_pred) ** 2) / total_variance)

# Output the results
print(f"Model: Simple Linear Regression")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R-squared: {r2:.4f}")
```
